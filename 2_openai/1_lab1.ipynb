```python
# The imports

import os
from dotenv import load_dotenv
from openai import AsyncOpenAI
from agents import Agent, Runner, trace, OpenAIChatCompletionsModel, set_tracing_disabled


```

```python
# The usual starting point

load_dotenv(override=True)

# No OpenAI key, so use Groq through its OpenAI-compatible API
groq_client = AsyncOpenAI(
    base_url="https://api.groq.com/openai/v1",
    api_key=os.getenv("GROQ_API_KEY"),
)

# Tracing uploads to platform.openai.com and needs an OpenAI key — disable it
set_tracing_disabled(True)

```

```python

# Make an agent with name, instructions, model

agent = Agent(
    name="Jokester",
    instructions="You are a joke teller",
    model=OpenAIChatCompletionsModel(model="llama-3.3-70b-versatile", openai_client=groq_client),
)
```

```python
# Run the joke with Runner.run(agent, prompt) then print final_output

with trace("Telling a joke"):
    result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents")
    print(result.final_output)
```
